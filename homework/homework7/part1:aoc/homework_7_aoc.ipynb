```python
import pandas as pd
```

```python
file_path = 'AOC_recent_tweets.txt'
df = pd.read_json(file_path)
```

```python
df = df.set_index('id')
```

```python
df
```

```python
def time_in_hours(created_at):
    created_at = pd.to_datetime(created_at)
    decimal_hours = created_at.dt.hour + created_at.dt.minute / 60 + created_at.dt.second / 3600
    return decimal_hours
```

```python
df['hours'] = time_in_hours(df['created_at'])
```

```python
selected_columns = ['created_at', 'hours', 'full_text']
```

```python
df_selected = df[selected_columns]
```

```python
df_selected
```

```python
output_file_path = 'AOC_tweets_processed.csv'
df_selected.to_csv(output_file_path)
```
