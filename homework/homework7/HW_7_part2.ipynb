```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
planets = sns.load_dataset('planets')
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=planets, x='orbital_period', y='mass', hue='method', palette='bright', alpha=0.7)
plt.xscale('log')
plt.yscale('log')
plt.title('Orbital Period vs Mass of Exoplanets')
plt.xlabel('Orbital Period (days)')
plt.ylabel('Mass (Jupiter Mass)')
plt.legend(title='Discovery Method')
plt.show()
```

```python
grouped_data = planets.groupby(['year', 'method']).size().unstack(fill_value=0)
```

```python
plt.figure(figsize=(16, 6))
grouped_data.plot(kind='bar', stacked=True, colormap='plasma')
plt.title('Number of Exoplanets Discovered by Year')
plt.xlabel('Year')
plt.ylabel('Number of Exoplanets Discovered')
plt.legend(title='Discovery Method', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
